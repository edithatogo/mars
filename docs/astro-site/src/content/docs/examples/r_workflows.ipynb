# R Runtime Notebook

This notebook-equivalent captures the R binding replay workflow.

## Workflow

1. Load a validated `ModelSpec` with `load_model_spec(...)`.
2. Evaluate `design_matrix(...)` and `predict_model(...)`.
3. Use `fit_model(...)` when the Rust-backed training bridge is available.

## Validation

Run `Rscript tests/conformance.R` or `R CMD check --no-manual --as-cran`.
