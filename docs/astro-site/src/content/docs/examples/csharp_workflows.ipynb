# C# Runtime Notebook

This notebook-equivalent documents the native ABI smoke path and the C# runtime wrapper.

## Workflow

1. Load a validated `ModelSpec` with `Runtime.LoadModelSpec(...)`.
2. Evaluate `Runtime.DesignMatrix(...)` and `Runtime.Predict(...)`.
3. Fit through `Runtime.FitModel(...)` and replay the returned portable spec.
