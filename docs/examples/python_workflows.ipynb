# pymars Python Workflow Notebook

This notebook mirrors the main Python tutorial: fit, predict, export, validate, and inspect.

```python
import numpy as np
import pymars as earth

X = np.array([[0.0, 0.1, 0.2], [0.2, 0.3, 0.4], [0.4, 0.1, 0.3]], dtype=float)
y = np.array([1.0, 1.8, 2.4], dtype=float)
model = earth.Earth(max_terms=6, max_degree=1)
model.fit(X, y)
predictions = model.predict(X)
spec = model.export_model()
summary = earth.inspect(spec)
validated = earth.validate(spec)
design = earth.design_matrix(spec, X)
predictions.tolist(), summary['model_type'], validated['spec_version'], design.shape
```
